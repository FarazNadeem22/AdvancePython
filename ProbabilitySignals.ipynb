```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
p = 0.5  # Probability of transmitting 0
sigma = 1.0  # Standard deviation of noise
a = -sigma**2 / 2 * np.log(p / (1 - p))  # Optimal decision threshold

# Generate x values
x = np.linspace(-4, 4, 1000)

# PDF of noise centered at -1 and +1
pdf_neg1 = norm.pdf(x, -1, sigma)
pdf_pos1 = norm.pdf(x, 1, sigma)

# Plot the distributions
plt.figure(figsize=(10, 6))
plt.plot(x, pdf_neg1, label="Transmitted: -1 (Received signal)", color='blue')
plt.plot(x, pdf_pos1, label="Transmitted: +1 (Received signal)", color='green')

# Mark the decision threshold
plt.axvline(x=a, color='red', linestyle='--', label=f'Decision Threshold (a = {a:.2f})')

# Fill the error regions
plt.fill_between(x, 0, pdf_neg1, where=(x >= a), color='blue', alpha=0.3, label='False Positive')
plt.fill_between(x, 0, pdf_pos1, where=(x < a), color='green', alpha=0.3, label='False Negative')

# Add labels and legend
plt.xlabel("Received Signal (R)")
plt.ylabel("Probability Density")
plt.title("Received Signal Distributions with Decision Threshold")
plt.legend()
plt.grid(True)
plt.show()

```
